```python
# importing the libraries
from bs4 import BeautifulSoup
import requests
import re
import time
import datetime
import pandas as pd
```

```python
def scrap_day(d):
    url="https://rpp.pe/archivo/" + str(d)

    # Make a GET request to fetch the raw HTML content
    html_content = requests.get(url).text

    # Parse the html content
    soup = BeautifulSoup(html_content, "lxml")

    list2 = soup.find_all(attrs={'class' : 'top-info'})
    list1 = soup.find_all(attrs={'class' : 'cont'})

    categories = []
    for i in list2:
        j = i.find('a')
        if("<img" in j):
            categories.append("-")
        else:
            categories.append(re.sub(r"\<[^<>]*\>", "", str(j)))

    description= []
    heading = []
    url = []
    for i in list1:
        res = i.find('p')
        des = re.sub(r"\<[^<>]*\>", "", str(res))
        description.append(str(des))

        res1 = i.find('h2')
        if(res1):
            splitted = str(res1).split(">")
            #print(splitted)
            url.append(splitted[1].replace("<a href=","").replace('"',''))
            heading.append(splitted[2].replace("</a",""))
    '''
    for i in range(len(categories)):
        print("------")
        print(categories[i])
        print(url[i])
        print(heading[i])
        print(description[i])
    '''
    df = pd.DataFrame(list(zip(categories, heading, description,url)),columns =['Categoría', 'Titular', 'Descripción', 'URL'])
    df['Fecha'] =  str(d)
    return df
```

```python
def scrap_dates(date1,date2):
    
    n_days = (date2-date1).days
    print(n_days)
    frames = []
    for i in range (0,n_days):
        print(scrap_day(date1+ datetime.timedelta(i)).shape)
        frames.append(scrap_day(date1+ datetime.timedelta(i)))
    
    result = pd.concat(frames)
    return result        
   
```

```python
def download(inicio, fin):
    
    news = scrap_dates(inicio,fin)
    
    file_name = str(inicio) + "_" + str(fin) + ".xlsx"
    news.to_excel(file_name) 
    return(news)
    
```

```python
inicio = datetime.date(2017,1,1)
fin = datetime.date(2018,1,1)
res = download(inicio,fin)
```

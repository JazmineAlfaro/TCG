```python
from bs4 import BeautifulSoup
import multiprocessing as mp
import time
import datetime
import pandas as pd
from multiprocessing import Pool
import news
```

```python
def scraping(date, date2):
    all_news = []
    
    n_days = (date2-date).days + 1
    print("Days: ", n_days)
    
    num_processors = mp.cpu_count()  
    print("Processors: ",num_processors)
    
    p=Pool(processes = num_processors)
    all_news =  p.map(news.scrap_page1,[str(date + datetime.timedelta(days=i)) for i in range(0,n_days)])
  
    return(all_news)
```

```python
def export(df1, name):
    news = pd.DataFrame()
    for i in df1:
        df = pd.DataFrame(i)
        df = df.transpose()
        news = pd.concat([news, df], axis=0)
    news.to_excel(name, encoding='utf-8-sig')
    return news
```

```python
def main(d1,d2):
    name = str(d1) + " " + str(d2)+ ".xlsx"
    res = export(scraping(d1,d2), name)
    return res
```

```python
start = time.time()
inicio = datetime.date(2018, 5, 1)
fin = datetime.date(2018, 8, 1)
resu = main(inicio,fin)
end = time.time()
print(end - start)
```
